```python
# Standard data science imports
import numpy as np
import pandas as pd
from pandas import DataFrame

# Visualization libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Statistics packages
import pylab
import statsmodels.api as sm
import statistics as stat
from scipy import stats
```

```python
data = pd.read_csv ('C:/Users/blien/Documents/WGU/D207/Medical_data/medical_clean.csv',dtype={'locationid':np.int64}) 
data.head()
```

```python
#renaming columns to better describe them
data.rename(columns = {'Item1':'Timely admission',
'Item2':'Timely treatment',
'Item3':'Timely visits',
'Item4':'Reliability',
'Item5':'Options',
'Item6':'Hours of treatment',
'Item7':'Courteous staff',
'Item8':'Active listening from Doctor'},
inplace=True)
data.head()
```

```python
#creating a new df for variables in research question
df = data[['VitD_levels', 'Gender']].copy()
print(df)
```

```python
#viewing data types of variables
df.dtypes
```

```python
#Grouping variables by gender
s = df.groupby('Gender')['VitD_levels'].apply(list)
s.apply(pd.Series).T
```

```python
#printing the means for each of our variables
print(('Male:        ') + str(int(stat.mean(s['Male']))))
print(('Female:      ') + str(int(stat.mean(s['Female']))))
print(('Nonbinary:   ') + str(int(stat.mean(s['Nonbinary']))))
```

```python
#running a t test to get the p value for our variables
stats.ttest_ind(s['Male'], s['Female'])
```

```python
#graphing the variables
sns.distplot(s['Male'], label = 'Male')
sns.distplot(s['Female'], label = 'Female')
sns.distplot(s['Nonbinary'], label = 'Nonbinary')
plt.legend()
```

```python
#printing f value and p value
f, p = stats.f_oneway(s['Male'],s['Female'],s['Nonbinary'])
print(f'f-value: {f}')
print(f'p-value: {p}')
```
